```python
# You are not allowed to import other packages.

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
```

### Part1:
Consider the dataset which has been given to you. Split the dataset into the `training dataset` and `test dataset`. You can use `80%` of the data for training and the rest for testing.

```python
dataset = pd.read_csv("./data/data01.csv")
print(dataset.describe())
dataset = dataset.to_numpy()
dataset[:, 0] = 1
X = dataset[:, :-1]
y = dataset[:, -1]

# implement the split part
# name them: x_train, x_test, y_train, y_test

n_samples = len(y)
n_train = int(n_samples * 0.8)

x_train = X[:n_train, :]
y_train = y[:n_train].reshape(-1, 1)

x_test = X[n_train:, :]
y_test = y[n_train:].reshape(-1, 1)
```

### Part2:
Assume there is simple linear relationship between X and Y. Train a simple `Linear Regression` on the training dataset and report its accuracy on the training and testing dataset, separately.

```python
# here you can implement via sgd or use least square solution. we suggest the second one inv(A^T A) * A^T y.
# consider m as the returned line --> m[0] = the slope , m[1] = y intercept, also don't forget to use numpy for calculating m.

w_star = np.linalg.pinv(x_train) @ y_train

m = (w_star[1,0], w_star[0,0])
x_train = x_train[:, 1:]
x_test = x_test[:, 1:]
```

```python
predicted_train_y = m[0] * x_train + m[1]
predicted_test_y = m[0] * x_test + m[1]

train_error = np.mean((predicted_train_y - y_train) ** 2)
test_error = np.mean((predicted_test_y - y_test) ** 2)

print(f"MSE train error: {train_error}\nMSE test error: {test_error}")
```

### Part5:
Now, consider the `Locally Weighted Linear Regression` approach for this problem. Do expect it performs better than simple `Linear Regression` for this dataset?

Yes, because the locally weighted linear regression considers only the data points in a neighborhood of the query point and uses those to fit a straight line to data (this is true for rectangular weight function, and in case of other weight functions, the weight of points decreases as we get farther from the query point). In this way, it estimates the curve with piecewise linear function (the length of segments can be controlled with a hyperparameter) and can capture the non-linearity of the data.

### Part6:
Here, we want to implement the `suggested model`. You can implement any weight function you want, the below implementation is just a sample:
<br><br/>
\begin{equation*}
    w^{(i)}=exp\left(-\frac{(\mathbf x^{(i)}-\mathbf x)^\top(\mathbf x^{(i)}-\mathbf x)}{2\tau^2}\right)
\end{equation*}

```python
def get_weight_matrix(the_query_point, train_dataset, hyperparameter):
    l = train_dataset.shape[0]
    W = np.mat(np.eye(l))
    for i, xi in enumerate(train_dataset):
        W[i, i] = np.exp((-np.linalg.norm(the_query_point - xi) ** 2)/(2 * hyperparameter ** 2))
    return W
```

Now implement the `closed form solution` based on the following formula:
<br><br/>
\begin{equation*}
    \boldsymbol\theta=(X^\top WX)^{-1}(X^\top WY)
\end{equation*}
<br><br/>
Of course if you use another form, implement your own.

```python
def predict(train_X, train_Y, query_x, hyperparameter):
    l = train_X.shape[0]
    X = np.hstack((train_X, np.ones((l, 1))))
    qx = np.array([query_x, 1], dtype=object)
    W = get_weight_matrix(qx, X, hyperparameter)

    theta = np.linalg.pinv(np.sqrt(W) @ X) @ (np.sqrt(W) @ train_Y)

    return np.dot(qx, theta)
```

### Part7:
Report accuracy of this model on the test dataset. Also, draw a plot that contains (x_test, y_test) pairs and (x_test, predicted_y) pairs that differ in color.

```python
hyperparameter = 0.08
predicted_test = np.array([predict(x_train, y_train, query, hyperparameter) for query in x_test])
error = np.mean((predicted_test.flatten() - y_test.flatten()) ** 2)
print(f"MSE test error: {error}")
```

```python
plt.scatter(x_test, y_test, c="blue")
plt.scatter(x_test, predicted_test, c="orange")
```

### Part9:
Now, we want to consider the `KNN` approach for this problem; Do you think it will perform well?

KNN in some sense, is like the locally weighted linear regression, since it uses points in the neighborhood of the query point to estimate the label for the query point. So, it's expected to work almost as well as locally weighted linear regression on the given dataset.

### Part10:
Implementing the `KNN model`.

```python
class KNN:
    def __init__(self, number_neighbours, X, Y):
        self.k = number_neighbours
        self.X = X
        self.Y = Y

    def distance(self, x1, x2):
        distance = np.linalg.norm(x1 - x2) # implement the distance function
        return distance

    def return_nn_ids(self, x):
        distances = np.array([self.distance(x, y) for y in self.X])
        indices = np.argpartition(distances, self.k)[:self.k]
        return indices

    def predict(self, x):
        nearest_indices = self.return_nn_ids(x)
        predicted_y = np.mean(self.Y[nearest_indices])
        return predicted_y
```

```python
def evaluation(k, x_train, y_train, x_test, y_test):
    model = KNN(k, x_train, y_train)
    predicted_y = np.array([model.predict(x) for x in x_test])
    error = np.mean((predicted_y.flatten() - y_test.flatten()) ** 2)
    return error, predicted_y
```

```python
k_vals = np.arange(1, 50)
errors = []
kf = KFold(n_splits=5)
for k in k_vals:
    mean_error = 0
    for train_idx, test_idx in kf.split(x_train):
        error, predicted_y = evaluation(k, x_train[train_idx, :], y_train[train_idx, :],
                                            x_train[test_idx, :], y_train[test_idx, :])
        mean_error += error
    
    errors.append(mean_error)

optimal_k = k_vals[np.argmin(errors)]
```

```python
k = optimal_k
error_knn, predicted_y = evaluation(k, x_train, y_train, x_test, y_test)
print(f"MSE error of KNN for test dataset: {error_knn}")
```

```python
plt.scatter(x_test, y_test, c="blue")
plt.scatter(x_test, predicted_y, c="orange")
```
